```python
from scipy.stats import beta
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)

a_array = [2, 5]
b_array = [2, 5]
lst = ['-', '--', '-.', '']
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, beta.pdf(x, 2, 2), lw = 1, alpha = 0.6, label = 'a = 2' + ',b = 2', ls = '-', linewidth = 4.0)
plt.plot(x, beta.pdf(x, 2, 5), lw = 1, alpha = 0.6, label = 'a = 2' + ',b = 5', ls = '--',linewidth = 4.0)
#plt.plot(x, beta.pdf(x, 5, 2), lw = 1, alpha = 0.6, label = 'a = 5' + ',b = 2', ls = '-.', linewidth = 4.0)
plt.plot(x, beta.pdf(x, 5, 2), lw = 1, alpha = 0.6, label = 'a = 5' + ',b = 2', ls = (0, (3, 1, 1, 1)), linewidth = 4.0)

plt.plot(x, beta.pdf(x, 0.5, 0.5), lw = 1, alpha = 0.6, label = 'a = 0.5' + ',b = 0.5', ls = ':', linewidth = 4.0)

plt.legend(frameon = False)
plt.show()
```

```python
for i, a in enumerate(a_array):
    for j, b in enumerate(b_array):
        plt.plot(x, beta.pdf(x, a, b), lw = 1, alpha = 0.6)
        
plt.show()
```
